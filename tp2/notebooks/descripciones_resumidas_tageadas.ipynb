```python
import pandas as pd

from summa import keywords
from summa import summarizer
from rake_nltk import Rake
```

```python
# Contar cantidad de signos de admiracion
```

```python
ruta_data = "../data"
ruta_train = ruta_data + "/train.csv"
ruta_test = ruta_data + "/test.csv"
```

```python
desc_train = pd.read_csv(f"{ruta_data}/descripciones_train_limpias.csv")
```

```python
desc_test = pd.read_csv(f"{ruta_data}/descripciones_test_limpias.csv")
```

```python
desc_train = desc_train.fillna("")
desc_test = desc_test.fillna("")
```

```python
# Workarounds a bugs de summa
desc_train["descripcion"] = desc_train["descripcion"].apply(lambda x: x.replace("urgue", "urge"))
desc_test["descripcion"] = desc_test["descripcion"].apply(lambda x: x.replace("urgue", "urge"))

desc_train["descripcion"] = desc_train["descripcion"].apply(lambda x: x.replace("pgue", "page"))
desc_test["descripcion"] = desc_test["descripcion"].apply(lambda x: x.replace("pgue", "page"))
```

```python
desc_train["descripcion"] = desc_train["descripcion"].apply(lambda x: summarizer.summarize(x, language="spanish", ratio=0.2))
desc_test["descripcion"] = desc_test["descripcion"].apply(lambda x: summarizer.summarize(x, language="spanish", ratio=0.2))
```
