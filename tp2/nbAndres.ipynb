```python
'''
reyenar metros cuadrados totales: teniendo su precio y el valor de metro cuadrado total
    promedio segun tipo de propiedad y provinciam,calculo los metros cuadrados totales
    que deberia tener. pag 19
metros cuadrados cubiertos: idem arriba pero con el grafico de la pagina 20

la mejor opcion porque hay alta correlacion entre los metros y el precio

                    
antiguedad: segun el tipo de propiedad

importante porque hay una leve relacion entre el precio y la antiguedad
'''
```

```python
train['precioMetroCuadrado'] = train['precio']/train['metrostotales']
train.head()
```

```python
metroCuadradoProv=train.groupby(['provincia','tipodepropiedad']).agg({'precioMetroCuadrado':'mean'}).reset_index()
metroCuadradoProv.head()
```

```python
metroCuadradoProv[(metroCuadradoProv['provincia']=='Distrito Federal') & (metroCuadradoProv['tipodepropiedad']=='Apartamento')].head()
```

```python
train=train.drop(['precioMetroCuadrado'], axis=1)
```

```python
train=train.merge(metroCuadradoProv,on=['provincia','tipodepropiedad'],how='inner')
```

```python
train.head()
```

```python
train['metrostotales'] = train['metrostotales'].fillna(train['precio']/train['precioMetroCuadrado'])
train.head()
```

```python
train.isnull().sum().head(23)
```

```python
'''

metros cuadrados cubiertos: idem arriba pero con el grafico de la pagina 20

la mejor opcion porque hay alta correlacion entre los metros y el precio

                    
antiguedad: segun el tipo de propiedad

importante porque hay una leve relacion entre el precio y la antiguedad

'''
```

```python
train['precioMetroCubierto'] = train['precio']/train['metroscubiertos']
```

```python
metroCuadradoProv=train.groupby(['provincia','tipodepropiedad']).agg({'precioMetroCubierto':'mean'}).reset_index()
metroCuadradoProv.head()
```

```python
train=train.drop(['precioMetroCubierto'], axis=1)
```

```python
train=train.merge(metroCuadradoProv,on=['provincia','tipodepropiedad'],how='inner')
```

```python
train['metroscubiertos'] = train['metroscubiertos'].fillna(train['precio']/train['precioMetroCubierto'])
```
